```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download the VADER lexicon if needed
nltk.download('vader_lexicon')

# Load the dataset
file_path = 'the final dataset.xlsx'  # Update with your file path
excel_data = pd.ExcelFile(file_path)
comments_data = excel_data.parse('Sheet2')

# Initialize VADER Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Define attributes to analyze
attributes = ["performance", "battery", "screen", "price", "brand", "camera"]

# Initialize dictionary to store sentiment counts for each attribute
attribute_counts = {attr: {"positive": 0, "neutral": 0, "negative": 0} for attr in attributes}

# Analyze each comment
for comment in comments_data['Comments'].dropna():
    sentiment_scores = sia.polarity_scores(comment)
    compound_score = sentiment_scores['compound']
    
    # Classify sentiment based on compound score
    if compound_score >= 0.05:
        sentiment = "positive"
    elif compound_score <= -0.05:
        sentiment = "negative"
    else:
        sentiment = "neutral"
    
    # Count mentions of each attribute
    for attr in attributes:
        if attr in comment.lower():
            attribute_counts[attr][sentiment] += 1

# Convert results to DataFrame
attribute_counts_df = pd.DataFrame(attribute_counts).T
attribute_counts_df.columns = ['Positive', 'Neutral', 'Negative']
print(attribute_counts_df)
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download the VADER lexicon if needed
nltk.download('vader_lexicon')

# Load your dataset (update the path as needed)
file_path = 'the final dataset.xlsx'
excel_data = pd.ExcelFile(file_path)
comments_data = excel_data.parse('Sheet2')

# Initialize VADER Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Initialize counters for overall sentiment
overall_counts = {"positive": 0, "neutral": 0, "negative": 0}

# Iterate through each comment and determine its sentiment
for comment in comments_data['Comments'].dropna():
    sentiment_scores = sia.polarity_scores(comment)
    compound_score = sentiment_scores['compound']
    
    if compound_score >= 0.05:
        overall_counts["positive"] += 1
    elif compound_score <= -0.05:
        overall_counts["negative"] += 1
    else:
        overall_counts["neutral"] += 1

print("Overall Sentiment Counts:", overall_counts)
```
