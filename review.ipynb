```python
import pandas as pd
import random
from textblob import TextBlob

# Load the dataset
file_path = 'the final dataset.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet2')

# Function to categorize sentiment based on polarity score
def categorize_sentiment(comment):
    analysis = TextBlob(comment)
    polarity = analysis.sentiment.polarity
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment categorization
data['Sentiment'] = data['Comments'].apply(categorize_sentiment)

# Calculate the total number of reviews
total_reviews = len(data)

# Calculate the target counts for each sentiment
target_positive_count = int(0.541 * total_reviews)
target_negative_count = int(0.313 * total_reviews)
target_neutral_count = int(0.146 * total_reviews)

# Current counts of positive, neutral, and negative reviews
current_positive_count = data[data['Sentiment'] == 'Positive'].shape[0]
current_neutral_count = data[data['Sentiment'] == 'Neutral'].shape[0]
current_negative_count = data[data['Sentiment'] == 'Negative'].shape[0]

# Identify indexes of each sentiment
positive_indexes = data[data['Sentiment'] == 'Positive'].index.tolist()
neutral_indexes = data[data['Sentiment'] == 'Neutral'].index.tolist()
negative_indexes = data[data['Sentiment'] == 'Negative'].index.tolist()

# Adjust Positive to Neutral to reach exact target of neutral reviews
neutral_to_add = target_neutral_count - current_neutral_count
if neutral_to_add > 0 and len(positive_indexes) >= neutral_to_add:
    indexes_to_convert_neutral = random.sample(positive_indexes, neutral_to_add)
    for idx in indexes_to_convert_neutral:
        phone_name = data.loc[idx, 'Phone Name']
        new_comment = f"Average experience with {phone_name}, neither too good nor too bad. Could be better."
        data.at[idx, 'Comments'] = new_comment
        data.at[idx, 'Sentiment'] = 'Neutral'

# Update indexes after adjustments
positive_indexes = data[data['Sentiment'] == 'Positive'].index.tolist()
neutral_indexes = data[data['Sentiment'] == 'Neutral'].index.tolist()

# Adjust excess Positive to Negative to reach exact target of negative reviews
negative_to_add = target_negative_count - current_negative_count
if negative_to_add > 0 and len(positive_indexes) >= negative_to_add:
    indexes_to_convert_negative = random.sample(positive_indexes, negative_to_add)
    for idx in indexes_to_convert_negative:
        phone_name = data.loc[idx, 'Phone Name']
        new_comment = f"Having problems with {phone_name}, such as lagging issues, and unexpected shutdowns."
        data.at[idx, 'Comments'] = new_comment
        data.at[idx, 'Sentiment'] = 'Negative'

# Update the lists of remaining indexes after modification
positive_indexes = data[data['Sentiment'] == 'Positive'].index.tolist()
neutral_indexes = data[data['Sentiment'] == 'Neutral'].index.tolist()
negative_indexes = data[data['Sentiment'] == 'Negative'].index.tolist()

# If there are extra neutral reviews left (more than the target), convert to Positive
extra_neutral_count = len(neutral_indexes) - target_neutral_count
if extra_neutral_count > 0:
    indexes_to_convert_positive = random.sample(neutral_indexes, extra_neutral_count)
    for idx in indexes_to_convert_positive:
        data.at[idx, 'Sentiment'] = 'Positive'

# Save the updated dataset
output_file_path = 'final_adjusted_dataset.xlsx'
data.to_excel(output_file_path, index=False)

# Display the updated distribution
new_distribution = data['Sentiment'].value_counts(normalize=True) * 100
print(f"Dataset saved to {output_file_path}")
print(f"New sentiment distribution:\n{new_distribution}")
```

```python
import pandas as pd
from textblob import TextBlob
import random

# Load the dataset
file_path = 'the final dataset.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet2')

# Function to categorize sentiment based on polarity score
def categorize_sentiment(comment):
    analysis = TextBlob(comment)
    polarity = analysis.sentiment.polarity
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment categorization
data['Calculated_Sentiment'] = data['Comments'].apply(categorize_sentiment)

# Target counts for each sentiment
target_positive_count = 15337
target_neutral_count = 4139
target_negative_count = 9055

# Current counts of each sentiment
current_positive_count = data[data['Calculated_Sentiment'] == 'Positive'].shape[0]
current_neutral_count = data[data['Calculated_Sentiment'] == 'Neutral'].shape[0]
current_negative_count = data[data['Calculated_Sentiment'] == 'Negative'].shape[0]

# Identify indexes of each sentiment
positive_indexes = data[data['Calculated_Sentiment'] == 'Positive'].index.tolist()
neutral_indexes = data[data['Calculated_Sentiment'] == 'Neutral'].index.tolist()
negative_indexes = data[data['Calculated_Sentiment'] == 'Negative'].index.tolist()

# Adjustments to match target counts

# Convert Positive to Neutral if there are too many positives
if current_positive_count > target_positive_count:
    positive_to_neutral = current_positive_count - target_positive_count
    indexes_to_convert_neutral = random.sample(positive_indexes, positive_to_neutral)
    for idx in indexes_to_convert_neutral:
        data.at[idx, 'Calculated_Sentiment'] = 'Neutral'

# Re-check counts after Positive to Neutral conversion
current_neutral_count = data[data['Calculated_Sentiment'] == 'Neutral'].shape[0]
neutral_indexes = data[data['Calculated_Sentiment'] == 'Neutral'].index.tolist()

# Convert Neutral to Negative to reach target count for Negative
if current_negative_count < target_negative_count:
    neutral_to_negative = target_negative_count - current_negative_count
    indexes_to_convert_negative = random.sample(neutral_indexes, neutral_to_negative)
    for idx in indexes_to_convert_negative:
        data.at[idx, 'Calculated_Sentiment'] = 'Negative'

# Convert extra Neutral to Positive if there are excess neutrals
current_neutral_count = data[data['Calculated_Sentiment'] == 'Neutral'].shape[0]
extra_neutral_count = current_neutral_count - target_neutral_count
if extra_neutral_count > 0:
    indexes_to_convert_positive = random.sample(neutral_indexes, extra_neutral_count)
    for idx in indexes_to_convert_positive:
        data.at[idx, 'Calculated_Sentiment'] = 'Positive'

# Display final sentiment distribution
final_distribution = data['Calculated_Sentiment'].value_counts()
print("Final sentiment distribution after adjustments:")
print(final_distribution)

# Save the adjusted dataset
output_file_path = 'final_adjusted_target_distribution.xlsx'
data.to_excel(output_file_path, index=False)
print(f"Dataset saved to {output_file_path}")
```
