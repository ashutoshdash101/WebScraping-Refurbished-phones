```python
import pandas as pd
import random

# Sample data (you can modify or extend this list based on your dataset)
data = [
    [10272], [7750], [6008], [5836], [5521], [5157], [3718], [3012], [2980], [2972],
    [2927], [2874], [2790], [2750], [2723], [2682], [2669], [2633], [2628], [2618]
]

# Create the dataframe
df = pd.DataFrame(data, columns=['f'])

# Helper function to generate random values and ensure they sum up to 'f'
def distribute_values(total):
    # Generate random percentages such that they sum to 100%
    percent_4_5 = random.uniform(40, 70)
    percent_1_2 = random.uniform(10, 40)
    percent_3 = 100 - percent_4_5 - percent_1_2

    # Calculate the values based on the percentages
    f_4_5 = int(total * (percent_4_5 / 100))
    f_1_2 = int(total * (percent_1_2 / 100))
    f_3 = total - f_4_5 - f_1_2

    # Calculate the actual percentages
    percent_4_5_actual = (f_4_5 / total) * 100
    percent_1_2_actual = (f_1_2 / total) * 100
    percent_3_actual = (f_3 / total) * 100

    return f_4_5, percent_4_5_actual, f_1_2, percent_1_2_actual, f_3, percent_3_actual

# Apply the distribution function to each row
df[['f_4,5', '%_4,5', 'f_1,2', '%_1,2', 'f_3', '%_3']] = df.apply(
    lambda row: pd.Series(distribute_values(row['f'])), axis=1
)

# Display the resulting dataframe
print(df)

# Save the dataframe to an Excel file
output_path = 'adjusted_table.xlsx'
df.to_excel(output_path, index=False)
print(f"Table saved to {output_path}")
```
